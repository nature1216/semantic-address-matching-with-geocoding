```python
from konlpy.tag import Kkma

# 꼬꼬마 형태소 분석기 객체 생성
kkma = Kkma()

text = "공대 1호관에 빨리 가고 싶다면 정문 앞 지름길을 이용하세요. 공대 1호관 정문 앞 지름길은 흙 길에 있습니다. 공대 1호관 정문 바로 앞에 있는 지름길은 시간이 없을 때 유용해요. 공대 1호관 정문 옆 지름길로 가면 공대 1호관까지 금방이에요."

# 형태소 추출
morphs = kkma.morphs(text) # text 변수에 저장된 문장을 형태소 단위로 토크나이징 합니다.
print(morphs)

# 형태소와 품사 태그 추출
pos = kkma.pos(text) # text 변수에 저장된 문장을 품사 태깅 합니다.
print(pos)

# 명사만 추출
nouns = kkma.nouns(text) # text 변수에 저장된 문장에서 품사가 명사인 단어들만 추출합니다.
print(nouns)

# 문장 분리
sentences = "공대 1호관에 빨리 가고 싶다면 정문 앞 지름길을 이용하세요. 공대 1호관 정문 앞 지름길은 흙 길에 있습니다. 공대 1호관 정문 바로 앞에 있는 지름길은 시간이 없을 때 유용해요. 공대 1호관 정문 옆 지름길로 가면 공대 1호관까지 금방이에요."
s = kkma.sentences(sentences) # 복합 문장이 있을 대 문장 단위로 토크나이징 합니다.
print(s)
```
